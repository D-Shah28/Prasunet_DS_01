```python
#importing neccesary library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
population_df = pd.read_csv('https://raw.githubusercontent.com/D-Shah28/Prasunet_DS_01/main/world_population.csv')
```

```python
# display the first 5 rows of dataset
population_df.head()
```

```python
#check the shape of the datset
population_df.shape
```

```python
#check the missing values
population_df.isnull().sum()
```

```python
# Count the number of countries in each continent
continent_counts = population_df['Continent'].value_counts()
print(continent_counts)
```

```python
# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(continent_counts.index, continent_counts.values)
plt.xlabel('Continent')
plt.ylabel('Number of Countries')
plt.title('Distribution of Continents')
for i, value in enumerate(continent_counts.values):
    plt.text(i, value, f'{value}', ha='center', va='bottom')

plt.show()
```

```python
# Create a histogram of population
plt.figure(figsize=(10, 6))
sns.histplot(population_df['2020 Population'], kde=True)
plt.xlabel('Population (2020)')
plt.ylabel('Frequency')
plt.title('Distribution of Population')
plt.show()
```

```python
# Create a boxplot of population growth rate
plt.figure(figsize=(10, 5))
sns.boxplot(y=population_df['Growth Rate'])
plt.xlabel('Population Growth Rate')
plt.ylabel('')
plt.title('Distribution of Population Growth Rate')
plt.show()
```

```python
#Bar chart of Top 10 Countries by Population
top_10_countries = population_df.nlargest(10, 'World Population Percentage')
plt.figure(figsize=(10, 6))
sns.barplot(x=top_10_countries['Country/Territory'], y=top_10_countries['World Population Percentage'])
plt.xlabel('Country')
plt.ylabel('Population')
plt.title('Top 10 Countries by Population')
plt.show()
```

A bar chart showing the distribution of countries across different continents.

A histogram showing the distribution of population across different countries.

A boxplot showing the distribution of population growth rates across different countries.

Bar chart off top 10 countries showing more than half of the world population is in India and China
